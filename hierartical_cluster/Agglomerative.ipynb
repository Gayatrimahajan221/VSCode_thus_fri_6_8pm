# Agglomerative clustering

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# read the dataset

import pandas as pd
df= pd.read_csv("Mall_Customers.csv")
df.head()
```

# categorize similar customer on basis of age , income, and speding score

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.columns= ['CustomerID', 'Gender', 'Age', 'Income', 'Spend']
```

```python
df.head()
```

seperate x data

```python
X = df[['Age', 'Income', 'Spend']]
X.head()
```

# preproccesing the x data

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

```python
X_pre.mean()
```

```python
X_pre.std()
```

# calculate optimal number of clustering on basis of silhoutee score

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
n = list(range(2,15))
n
```

```python
scores = []
for i in n:
    # Create the model
    model = AgglomerativeClustering(n_clusters=i)
    model.fit(X_pre)
    # Predict the model clusters
    lb = model.labels_
    # calculate silhoutee scores
    sil = silhouette_score(X_pre, lb)
    scores.append(sil)
```

```python
scores
```

# visualize silhoutee score

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(12, 6))
plt.plot(n, scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutee Score')
plt.title('Silhoutee Scores for various clusters')
plt.grid()
plt.show()
```

```python
# based on above graph for 6 cluster we haave highest score of 0.42

#create a agglomerative model with n_cluster=6
```

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(X_pre)
```

```python
model.labels_
```

```python
# save ablve result in x
```
