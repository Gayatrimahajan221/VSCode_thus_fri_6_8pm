```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# read dataset
```

```python
import pandas as pd
```

```python
df= pd.read_csv("AirPassengers.csv", index_col='Month', parse_dates=True)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.columns = ['nPass']
```

```python
df.head()
```

```python
# visualization time series data
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.plot(df['nPass'])
plt.xlabel('Months')
plt.ylabel('number of air passenger')
plt.title('air passenger ')
plt.grid()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
Q= seasonal_decompose(df['nPass'], model='multiplicative')
```

```python
Q.plot()
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(12, 6))
plt.plot(df['nPass'], label='orignal')
plt.plot(Q.trend, label='Trend')
plt.xlabel('Months')
plt.ylabel('Number of Air Passengers')
plt.title('Air Passengers 1949-1960')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(Q.seasonal)
plt.xlabel('Month')
plt.ylabel('Seasonality')
plt.title('Seasonality Plot')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(Q.resid)
plt.xlabel('Month')
plt.ylabel('Residual')
plt.title('Residual Plot')
plt.show()
```
